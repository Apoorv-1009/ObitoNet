```python
import os
import pandas as pd

# Load your CSV file
file_path = "../dataset/Dataset/labels.csv" # Replace with your file path
data = pd.read_csv(file_path)

# # Access the first column as a list
# first_column_list = data.iloc[:, 0].tolist()

# Function to check if paths exist
def check_paths_exist(column):
    # for path in column:
    #     print("./" + path)
    missing_paths = [path for path in column if not os.path.exists("../dataset/Dataset/" + path)]
    return missing_paths

# Iterate through each column and check if paths exist
for column_name in data.columns:
    print(f"Checking column: {column_name}")
    missing = check_paths_exist(data[column_name])
    if missing:
        print(f"Missing paths in {column_name}:")
        print(missing)
    else:
        print(f"All paths in {column_name} exist.")
```
